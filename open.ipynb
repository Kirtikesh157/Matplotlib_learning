```python
# Check Pallindrome 
ch =int(input("Enter the value"))
temp = ch
rev=0
while temp > 0:
    digit = temp % 10
    rev = rev * 10 + digit
    temp //= 10
    if rev==ch:
        print(f"the {ch} is pallindrome")
    else:
        print(f"the {ch} is not a pallindrom ")
```

```python
num = int(input("Enter a number: "))
temp = num
rev = 0

while temp > 0:
    digit = temp % 10       
    rev = rev * 10 + digit  
    temp //= 10             

if num == rev:
    print(f" Yes! {num} is a Palindrome number.")
else:
    print(f" No! {num} is NOT a Palindrome number.")
```
